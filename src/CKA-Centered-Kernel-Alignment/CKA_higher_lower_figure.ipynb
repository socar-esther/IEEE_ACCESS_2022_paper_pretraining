```python
import pickle
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## higher stage : naive

```python
# 1. upstream
with open('higher_lower_outputs/lower_upstream/naive.pth_naive.pth.pkl', 'rb') as f : 
    data = pickle.load(f)
naive_list = np.ravel(data).tolist()
weight_nm_list = ['naive'] * len(naive_list)
temp = {
    'weight_nm' : weight_nm_list, 
    'value' : naive_list
}
data = pd.DataFrame(temp)
data.head()
```

```python
# 2. 100
with open('higher_lower_outputs/lower_100/naive.pth_naive_acc_0.75483.pth.pkl', 'rb') as f : 
    data_2 = pickle.load(f)
naive_100_list = np.ravel(data_2).tolist()
weight_nm_list = ['naive_100'] * len(naive_100_list)
temp = {
    'weight_nm' : weight_nm_list, 
    'value' : naive_100_list
}
data_2 = pd.DataFrame(temp)
data_2.head()
```

```python
# 2. 50
with open('higher_lower_outputs/lower_50/naive.pth_naive_acc_0.73957.pth.pkl', 'rb') as f : 
    data_3 = pickle.load(f)
naive_50_list = np.ravel(data_3).tolist()
weight_nm_list = ['naive_50'] * len(naive_50_list)
temp = {
    'weight_nm' : weight_nm_list, 
    'value' : naive_50_list
}
data_3 = pd.DataFrame(temp)
data_3.head()
```

```python
# 3. 10
with open('higher_lower_outputs/lower_10/naive.pth_naive_acc_0.61750.pth.pkl', 'rb') as f : 
    data_4 = pickle.load(f)
naive_10_list = np.ravel(data_4).tolist()
weight_nm_list = ['naive_10'] * len(naive_10_list)
temp = {
    'weight_nm' : weight_nm_list, 
    'value' : naive_10_list
}
data_4 = pd.DataFrame(temp)
data_4.head()
```

```python
# 데이터 다 합치기
final = pd.concat([data, data_2, data_3, data_4])
final.head()
```

```python
final.shape
```

```python
# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 5)
plt.rcParams['font.size'] = 15

# 2. 데이터 준비

# 3. 그래프 그리기
fig, ax = plt.subplots()

violin = ax.violinplot([data['value'].tolist(), data_2['value'].tolist(), data_3['value'].tolist(), data_4['value'].tolist()], showmeans=True)
ax.set_ylim(0.4, 1.1)
ax.set_xticks(np.arange(1, 5))
plt.grid(True,color='gray', alpha=0.1)
ax.set_xticklabels(['Naive', 'Naive \n 100%', 'Naive \n 50%', 'Naive \n10%'])
ax.set_xlabel('Naive')
ax.set_ylabel('CKA similarity')

violin['bodies'][0].set_facecolor('blue')
violin['bodies'][1].set_facecolor('red')
violin['bodies'][2].set_facecolor('green')

violin['cbars'].set_edgecolor('gray')
violin['cmaxes'].set_edgecolor('gray')
violin['cmins'].set_edgecolor('gray')
violin['cmeans'].set_edgecolor('gray')

plt.show()
```
